## 기본 패키지 설치 확인 (1)

```python
from math import sin, pi
sin(pi/2.0)
```

- math 패키지의 sin함수 pi함수를 사용
- import에서 여러 함수 사용시 ','를 이용

```python
# from math import sin, pi
for k36_i in range(360):
    k36_sin_value = sin(k36_i * pi / 180)
    k36_space_value = int((1 + k36_sin_value) * 50)
    print(" " * k36_space_value, end = "")
    print("[%f]" % k36_sin_value)
```

- print()
    - end="" 옵션 : print 함수가 출력된 후 추가할 문자 지정, 줄을 바꾸지 않고 다음 줄에 있는 print값 출력
    - %f : 10진수 실수(float) 서식문자, 각 위치에 대응하는 변수의 값을 출력해준다.
        - %d(10진수 정수), %lf(실수-double형), %c(단일 문자), %s(문자열), %o(8진수 정수), %x(16진수 정수) 등
- range(n) : 0부터 n-1까지 숫자의 범위를 지정
- code : k36_i의 값은 0부터 359까지로 1씩 증가하는 동안 반복
    - k36_sin_value에 $\sin($k36_i * $\pi / 180)$ 값을 저장, k36_space_value값에는 (1 + k36_sin_value) * 50 한 값의 정수값을 저장
    - k36_space_value의 값만큼 여백을 출력하고 바로 [k36_sin_value값]을 출력한다. $\to$ sin함수의 그래프 형태로 값이 출력됨.

```python
import statistics as s
k36_a = [100, 95, 25, 33, 45, 90, 100, 35, 25]
print("입력", k36_a)

print("평균", s.mean(k36_a))      # mean
print("표준편차", s.stdev(k36_a)) # std
print("분산", s.variance(k36_a))  # var
```

- statistics 패키지를 가져와서 s로 사용
- k36_a라는 리스트([])를 생성해서 값을 저장
- statistics 패키지의 mean, stdev, variance 메소드를 사용해서 평균, 표준편차, 분산을 구한다.

```python
from math import *
from scipy.optimize import fsolve

def k36_f(k36_x):
    return k36_x - exp(-k36_x)

fsolve(k36_f, 0.5)
```

- 다음 방정식의 해를 구한다. $\to$ $x = e^{-x}$
- math 패키지의 모든 메소드 사용
- scipy.optimize 패키지의 fsolve 메소드 사용
- def~:를 사용해서 함수를 정의 $\to$ 풀고자 하는 방정식을 $f(x) = 0$ 형식으로 k36_f라는 함수 정의
- exp() : 지수 함수를 계산해서 반환하는 메소드
- fsolve(함수, 초기값) : 시작점이 초기값인 함수의 근을 찾는 메소드

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

- matplotlib.pyplot 패키지를 plt로 사용
- plot(x, y) : x와 y값을 사용하여 그래프를 생성하는 메소드
- show() : 생성된 그래프를 보여주는(화면에 출력하는) 메소드

```python
# from math import sin, pi
#import matplotlib.pyplot as plt

k36_graph_value = []
for k36_i in range(360):
    k36_graph_value.append(sin(k36_i * pi / 180))

plt.plot(k36_graph_value)
plt.show()
```

- 같은 파일에서 import as를 실행한 경우 다시 실행하지 않아도 사용 가능하다.
- k36_graph_value 라는 리스트를 생성하고 반복문(for)과 append 메서드를 사용하여 값을 추가한다.
- k36_i의 값은 0부터 359까지, 리스트 k36_graph_value에 $\sin($k36_i * $\pi / 180)$ 값을 추가
- 리스트 k36_graph_value의 그래프를 그리고 화면에 출력

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

def k36_barplot():
    plt.bar([1, 2, 3, 4, 5], [10, 13, 11, 20, 25], color = "red", width = 0.5, antialiased = True, fill = True)
    plt.xlabel("번호")
    plt.ylabel("점수")
    plt.title("우리반 성적")
    plt.legend(["국어", "영어"])
    plt.show()

k36_barplot()
```

- rcParams : matplotlib의 스타일 파라미터, rcParams에 정해진 기본값들을 수정해서 그래프의 스타일을 설정할 수 있다.
- 'font.family' : 설정되어 있는 글꼴, plt.rcParams['font.family'] = 'Malgun Gothic'를 이용해서 맑은 고딕으로 글꼴 설정
- plt.rcParams['axes.unicode_minus'] = False로 설정하면 한글 폰트로 설정했을 때의 '-' 기호가 깨지는 것을 방지할 수 있다.
- def~:를 사용해서 막대 그래프를 그리는 함수 k36_barplot을 생성
- bar() : 막대 그래프를 그리는 메서드
  - color = : 그래프의 색 지정
  - width = : 막대의 너비 조정
  - antialiased = True : 안티앨리어싱(=이미지의 가장자리에서 발생하는 계단 현상 완화하는 기술) 렌더링 사용 여부를 결정
  - fill = True : 막대 색 채우기 여부를 결정
  - xlabel/ylabel : x축/y축 이름 지정
  - title : 그래프의 제목 지정
  - legend : 그래프의 범례 지정
