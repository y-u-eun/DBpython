## 리스트 이용 pyplot 그래프 그리기 (1)

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

- matplotlib.pyplot 패키지를 plt로 사용
- plot(x, y) : x와 y의 값을 이용해 그래프 생성
- show() : 생성된 그래프를 보여준다. (화면에 출력)

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

- x, y 값 인자에 대해 선의 색상과 형태를 지정하는 포맷 문자열 (format string)을 세번쨰 인자에 입력할 수 있다.
- 포맷 문자열 'ro'는 빨간색(red)의 원형(o) 마커를 의미한다.
- 추가적으로 'b-'는 파란색(blue)의 실선(-)을 의미한다.
- matplotlit.pyplot 패키지의 axis() 함수를 이용하면 축의 범위를 지정할 수 있다. $\to$ [xmin,  xmax,  ymin,  ymax]

```python
# bar-chart
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

def barplot():
    plt.bar([1, 2, 3, 4, 5], [10, 13, 11, 20, 25], color = "red", width = 0.5, antialiased = True, fill = True)
    plt.xlabel("번호")
    plt.ylabel("점수")
    plt.title("우리반 성적")
    plt.legend(["국어", "영어"])
    plt.show()

barplot()
```

- rcParams : matplotlib의 스타일 파라미터, rcParams에 정해진 기본값들을 수정해서 그래프의 스타일을 설정할 수 있다.
- 'font.family' : 설정되어 있는 글꼴, plt.rcParams['font.family'] = 'Malgun Gothic'를 이용해서 맑은 고딕으로 글꼴 설정
- plt.rcParams['axes.unicode_minus'] = False로 설정하면 한글 폰트로 설정했을 때의 '-' 기호가 깨지는 것을 방지할 수 있다.
- def~:를 사용해서 막대 그래프를 그리는 함수 k36_barplot을 생성
- bar() : 막대 그래프를 그리는 메서드
  - color = : 그래프의 색 지정
  - width = : 막대의 너비 조정
  - antialiased =  : 안티앨리어싱(=이미지의 가장자리에서 발생하는 계단 현상 완화하는 기술) 렌더링 사용 여부를 결정
  - fill =  : 막대 색 채우기 여부를 결정
  - xlabel/ylabel : x축/y축 이름 지정
  - title : 그래프의 제목 지정
  - legend : 그래프의 범례 지정

## 리스트 이용 pyplot 그래프 그리기 (2)

- scipy.linalg 는 선형 대수 연산을 위한 패키지
- scipy는 거의 항상 numpy와 같이 사용 된다.

```python
import numpy as np
from scipy.linalg import *

arr = np.array([[1, 2], [3, 4]])
print("====")
print(arr)

print("====")
print(det(arr))

iarr = inv(arr)
print("====")
print(iarr)

iarr = inv(arr)
print("====")
print(np.dot(arr, iarr))
```

- np.array() 를 이용해서 행렬을 생성
- det() : 정사각 행렬의 행렬식을 반환하는 numpy의 함수
- inv() : 정사각 행렬의 역행렬을 반환하는 numpy의 함수
- dot() : 두 배열의 내적값 또는 행렬의 곱셈 결과를 반환하는 numpy의 함수

```python
%matplotlib inline
x = np.linspace(-1, 1, 101)
y = np.exp(-x)
plt.plot(x, y)
plt.plot(x, x)
plt.show()
```

- 다음 방정식 $x = e^{-x}$ 의 해를 구한다.
- %matplotlib inline : jupyter notebook 셀 안에서 그래프를 바로 볼 수 있도록 출력하는 명렁어
- linspace(start, end, num) : start부터 end까지 num개의 요소로 1차원 배열을 생성하는 함수
- $x$의 그래프(=주황색)과 $e^{-x}$의 그래프(=파란색)을 각각 그려보면 0.5 근처에서 해가 존재하는 것을 알 수 있다.

```python
# pi-chart
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

def piegraph():
    plt.pie([10, 20, 30, 40], labels = ["카리나", "윈터","지젤","닝닝"], shadow = False,
            colors = ['red', 'yellow', 'grey', 'magenta'], autopct = '%.1f%%')
    plt.xlabel("의미없음")
    plt.ylabel("의미없음")
    plt.title("에스파 짱")
    plt.legend(["카리나", "윈터","지젤","닝닝"])
    plt.show()

piegraph()
```

- rcParams : matplotlib의 스타일 파라미터, rcParams에 정해진 기본값들을 수정해서 그래프의 스타일을 설정할 수 있다.
- 'font.family' : 설정되어 있는 글꼴, plt.rcParams['font.family'] = 'Malgun Gothic'를 이용해서 맑은 고딕으로 글꼴 설정
- plt.rcParams['axes.unicode_minus'] = False로 설정하면 한글 폰트로 설정했을 때의 '-' 기호가 깨지는 것을 방지할 수 있다.
- def~:를 사용해서 원형 그래프를 그리는 함수 k36_piegraph를 생성
- pie() : 원형 그래프를 그리는 메서드
  - colors = : 그래프의 색 지정
  - labels = : 각 칸(=색)의 이름을 지정
  - shadow = : 그래프의 그림자 표시 여부를 결정
  - autopct =  : 퍼센트를 파이차트 안에 표시(%.1f는 소수점 첫째자리까지, %.2f는 소수점 둘째자리까지, %.1%%는 퍼센트 기호를 포함)
  - xlabel/ylabel : x축/y축 이름 지정
  - title : 그래프의 제목 지정
  - legend : 그래프의 범례 지정
